```python
import pandas as pd
from pathlib import Path
import ast
```

```python
def parse_missing_cutout_files(csv_file: str) -> dict:
    """Parse the CSV file to extract the lists of missing image, metadata, and mask names."""
    df = pd.read_csv(csv_file)

    # Convert the string representation of lists to actual lists
    df['MissingPng'] = df['MissingPng'].apply(ast.literal_eval)
    df['MissingJson'] = df['MissingJson'].apply(ast.literal_eval)
    df['MissingMask'] = df['MissingMask'].apply(ast.literal_eval)

    missing_files = {
        "missing_pngs": df['MissingPng'].tolist(),
        "missing_jsons": df['MissingJson'].tolist(),
        "missing_masks": df['MissingMask'].tolist()
    }

    return missing_files

def parse_missing_developed_files(csv_file: str) -> dict:
    """Parse the CSV file to extract the lists of missing image, metadata, and mask names."""
    df = pd.read_csv(csv_file)

    # Convert the string representation of lists to actual lists
    df['MissingImages'] = df['MissingImages'].apply(ast.literal_eval)
    df['MissingMetadata'] = df['MissingMetadata'].apply(ast.literal_eval)
    df['MissingMetaMasks'] = df['MissingMetaMasks'].apply(ast.literal_eval)

    missing_files = {
        "missing_images": df['MissingImages'].tolist(),
        "missing_metadata": df['MissingMetadata'].tolist(),
        "missing_meta_masks": df['MissingMetaMasks'].tolist()
    }

    return missing_files
```

```python
csv_file = "../semif_cutouts_batch_details_20240718_143448.csv"  # Replace with your actual CSV file path
# missing_files = parse_missing_developed_files(csv_file)
missing_files = parse_missing_cutout_files(csv_file)

print("Missing Pngs:")
for missing_list in missing_files["missing_pngs"]:
    print(missing_list)
    for missing_img in missing_list:
        print(missing_img + ".jpg")
```
